#### Question 1.
We consider the parameters $\theta = (\sigma,\sigma_\xi, \sigma_\tau,\bar{t_o},\bar{v_o},)$ and the complete data $(y,z)$ where $y = (y_1,...,y_N)$, $y_i = (y_{i1},...,y_{iK})$, $z=(z_{pop},(z_i)_{i=1,...N}$, $z_pop=(t_0, v_0)$ and $z_i=(\xi_i, \tau_i)$We begin by observing the fact that we can decompose the log-likelihood of the complete data as a sum of four terms:

$$
\log q(y,z,\theta) = \underbrace{\log q(y|z,\theta)}_\text{(1)} + \underbrace{\log q(z_{pop}|\theta)}_{(2)}+\underbrace{\log q((z_i)_{i=1}|\theta)}_{(3)}+ \underbrace{\log q(\theta)}_{(4)}
$$

We have the following:
For $(1)$, we observe that $y_{ij}\sim \mathcal{N}(d(t_{ij}),\sigma^2)$ and that we assume the N individuals to be independent and also the measurement for each individual.
$$
(1) = \sum_{i=1}^N \log(y_i|z_{pop},z_i, \theta) = \sum_{i=1}^N \sum_{i=1}^K \log(y_{ij}|z_{pop},z_i,\theta) =  -\frac{1}{2}\sum_{i=1}^N \sum_{i=1}^K \left(\frac{y_{ij}-d_i(t_{ij})}{\sigma} \right)^2 - NK \log(\sigma)-NK\log(\sqrt{2\pi})
$$

$$
(2) = \log q(t_0|\theta) + \log q(v_0|\theta) = -\frac{1}{2} \left( \frac{t_0-\bar{t_0}}{\sigma_{t_0}} \right)^2 - \log(\sigma_{t_0}) -\frac{1}{2} \left( \frac{v_0-\bar{v_0}}{\sigma_{v_0}} \right)^2 - \log(\sigma_{v_0}) -2\log(\sqrt{2\pi})
$$

$$
(3) = \sum_{i=1}^N \log(z_i| \theta) = \sum_{i=1}^N \log(\xi_i| \theta)+\log(\tau_i| \theta) = -\frac{1}{2}\sum_{i=1}^N \left[ \left( \frac{\xi_i}{\sigma_{\xi_i}} \right)^2 + \left( \frac{\tau_i}{\sigma_{\tau_i}}  \right)^2  \right] - N\log(\sigma_\xi) -N\log(\sigma_\tau) - 2N\log(\sqrt{2\pi}) 
$$

$$
(4) = \log q(\bar{t_0}) + \log q(\bar{v_0}) + \log q(\sigma)+ \log q(\sigma_\xi)+ \log q(\sigma_\tau) = -\frac{1}{2} \left( \frac{\bar{t_0}-\bar{\bar{t_0}}}{s_{t_0}} \right)^2 - \log(s_{t_0})-\frac{1}{2} \left( \frac{\bar{v_0}-\bar{\bar{v_0}}}{s_{v_0}} \right)^2 - \log(s_{v_0}) -2\log(\sqrt{2\pi}) -(m+2)\log(\sigma)-\frac{v^2}{2\sigma}- (m+2)\log(\sigma_\tau)-\frac{v^2}{2\sigma_\tau}- (m+2)\log(\sigma_\xi)-\frac{v^2}{2\xi_i} -3\log(\Gamma\left(\frac{m}{2}\right))+3m\log(v)
$$

Now, we need to prove that this is an exponential model, i.e., that it can be written as:
$$
\log q(y,z,\theta) = \phi_1(\theta) + \phi_2(y,z)+ \langle S(y,z), \psi(\theta) \rangle + cte
$$

It belongs to an exponential family. Indeed, developping the squares in $(2)$ and $(4)$, and defining $\phi_1(\theta)$, $S(\theta) = (S_1(\theta), S_2(\theta), S_3(\theta), S_4(\theta), S_5(\theta), S_6(\theta))$, $\psi(\theta) = (\psi_1(\theta), \psi_2(\theta), \psi_3(\theta), \psi_4(\theta), \psi_5(\theta), \psi_6(\theta))$ and $\phi_2(y,z)$ as:

$$
S_1(y,z) = \frac{1}{KN}\sum_{i=1}^N\sum_{j=1}^K(y_{ij}-d_i(t_{ij}))^2; \\
S_2(y,z) = \frac{1}{N}\sum_{i=1}^N\xi_i^2; \text{  }
S_3(y,z) =  \frac{1}{N}\sum_{i=1}^N\tau_i^2; \\
S_4(y,z) = t_0; \text{  }
S_5(y,z) = v_0; 
$$

$$
\psi_1(\theta) = -\frac{KN}{2\sigma^2}; \\
\psi_2(\theta) = -\frac{N}{2\sigma_\xi^2}; \text{  }
\psi_3(\theta) = -\frac{N}{2\sigma_\tau^2}; \\
\psi_4(\theta) = -\frac{\bar{t_0}}{\sigma_{t_0}^2}; \text{  }
\psi_5(\theta) = -\frac{\bar{v_0}}{\sigma_{v_0}^2};
$$

$$
\phi_1 (\theta) = -NK\log(\sigma)-N\log(\sigma_\xi)-N\log(\sigma_\tau)-(m+2)(\log(\sigma)+\log(\sigma_\xi)+\log(\sigma_\tau))-\frac{v^2}{2}(\frac{1}{\sigma^2}+\frac{1}{\sigma_\xi^2}+\frac{1}{\sigma_\tau^2})-\frac{\bar{t_0^2}}{2\sigma_{t0}^2}-\frac{\bar{v_0^2}}{2\sigma_{v0}^2}-\frac{\bar{t_0}^2-2\bar{t_0}\bar{\bar{t_0}}}{2s_{t_0}^2}-\frac{\bar{v_0}^2-2\bar{v_0}\bar{\bar{v_0}}}{2s_{v_0}^2}
$$

$$
\phi_2 (y, z) = -\frac{t_0^2}{\sigma_{t_0}^2} - \frac{v_0^2}{\sigma_{v_0}^2}
$$
the remaining terms belong to the constant since they do not depend on the parameters nor the variables.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import invwishart
from scipy.stats import multivariate_normal
from scipy.stats import norm
```

```python
#Global variables
m = 6
v = 2
t_bb_0 = 0.5
v_bb_0 = 1
sig_t0 = 0.1
sig_v0 = 0.1    
s_t0 = 0.1
s_v0= 0.1
p_0 = 0 #initial position

```

```python
def generate_data(N, k):
    '''
    Generates a sample of the model described in the exercise.
    
    :param N: number of individuals
    :param k: vector with the number of measurements for each of the N individuals
    
    :return y,theta, z: variable, parameters, and latent variable as described in the exercise
    '''

    
    # We sample the parameters theta using the prior distributions
    #We sample theta_pop (parameters of the fixed effects)
    t_b_0 = np.random.normal(t_bb_0, s_t0)
    v_b_0 = np.random.normal(v_bb_0, s_v0)
    
    #We sample theta_i (parameters of the random effects)
    sig_xi = invwishart.rvs(m, v)
    sig_tau = invwishart.rvs(m,v)
    sig = invwishart.rvs(m,v)
    
    
    #We sample the latent variables z_i and z_pop
    # Latent variable of the random effects
    xi = np.asarray([np.random.normal(0, sig_xi) for i in range(N)])
    alpha = np.exp(xi)
    tau = np.asarray([np.random.normal(0, sig_tau) for i in range(N)])

    #Latent variable fixed effects
    t_0 = np.random.normal(t_b_0, sig_t0)
    v_0 = np.random.normal(v_b_0, sig_v0)
    
  
    #For each individual we sample k measurments
    y=np.zeros([N, k])
    for i in range(N):
        #We take the time to be a sequence of 1 unit of difference and starting at time 0
        for j in range(k):
            y[i][j] = p_0 + v_0*alpha[i]*(time[j]-t_0-tau[i]) + np.random.normal(0,sig)
    
    theta = np.asarray([t_b_0, v_b_0, sig, sig_xi, sig_tau])
    
    z = np.asarray([t_0, v_0])
    print(z)
    
    for i,j in zip(xi, tau):
        z = np.append(z,[i])
        z = np.append(z,[j])

    return y, theta, z
```

### Question 3. Hasting Metropolitan

Using the expression of the log-likelihood as an exponential model, we obtain, using Bayes' rule:

$$
\alpha(z^*,z^{(k)}) = \frac{q(z^*|y,\theta)}{q(z^{(k)}|y,\theta)} = \frac{q(y,z^*|\theta) q(y|\theta)}{ q(y|\theta) q(y,z^{(k)}|\theta)} = frac{q(y,z^*|\theta)}{q(y,z^{(k)}|\theta)}
$$

$$
\implies \log(\alpha(z^*,z^{(k)})) = \log q(y,z^*|\theta)- \log q(y,z^{(k)}|\theta) = \underbrace{\phi_2(y,z^*)+ \langle S(y,z^*), \psi(\theta) \rangle }_\text{l(z^*)} - \underbrace{\phi_2(y,z^{(k)})+ \langle S(y,z^{(k)}), \psi(\theta) \rangle }_\text{l(z^{(k)})}
$$

$$
\implies \alpha(z^*,z^{(k)}) = e^{l(z^*)-l(z^{(k)})}
$$

```python
def aux_log_likelihood(data, theta, z):
    '''
    Computes the log-likelihood of the complete data without the terms that depend only on the parameters.
    
    :param data: values of y
    :param theta: parameters
    :param z: point in which we want to compute this function
    '''
    S = np.zeros([5])    
    N = data.shape[0] #number of individuals
    K = data.shape[1] #number of measurements per individual
                
    #We decompose the terms of z for simplicity
    z_pop = z[0:2] #fixed effects 
    t_0 = z_pop[0]
    v_0 = z_pop[1]
                 
    z_re = z[2:] #random effects
    n = int(z_re.shape[0]/2 )
    xi = np.asarray([z_re[2*k] for k in range(n)]) #xi
    tau = np.asarray([z_re[2*k+1] for k in range(n)]) #tau 
    
    #We decompose the parameters in theta for simplicity
    t_b_0 = theta[0]
    v_b_0 = theta[1]
    sig = theta[2]
    sig_xi = theta[3]
    sig_tau = theta[4]
    
    #We compute the terms as they appear in the exponential form of the likelihood
    phi = -((t_0/sig_t0)**2+(v_0/sig_v0)**2)*(1/2) 
    
    S = np.zeros(5)
    for i in range(N):
        for j in range(K):
            S[0] = S[0] + (data[i,j]-(p_0 + v_0*np.exp(xi[i])*(time[j]-t_0-tau[i])))**2
    S[0]=S[0]/(K*N)                 
    S[1]=(1/N)*((xi**2).sum())                         
    S[2] = (1/N)*((tau**2).sum())
    S[3] = t_0
    S[4] = v_0
            
    psi = np.zeros(5)
    psi[0] = -K*N/(2*(sig**2))
    psi[1] = -N/(2*(sig_xi**2))
    psi[2] = -N/(2*(sig_tau**2))
    psi[3] = t_b_0/(sig_t0**2)
    psi[4] = v_b_0/(sig_v0**2)
    
    return phi + np.dot(S,psi)
```

```python
def alpha_1(data,theta, z_prop, z):
    '''
    Computes the likelihood ratio between the proposed point and the previous sate of the Markov Chain
    in a H-M algorithm
    
    '''
    log_prop = aux_log_likelihood(data, theta, z_prop)
    log_z = aux_log_likelihood(data, theta, z)
    
    return np.exp(log_prop-log_z)
```

```python
def SRW_hasting_metropolis_z(data, z0,theta,sample_size,burnIn = 10000, sigma=0.001):
    '''
    Symmetric Random Walk Hasting Metropolis using multivariate gaussian distributions centered in the 
    the previous point as proposal distributions.
    
    :param data: values of y
    :param z0: initial point of the Markov Chain
    :param theta: parameters of the sample
    :param sample_size: size of the sample we want
    :burnIn: amount of samples to be discarded at the beginning of the Markov Chain
    :param sigma: covariance matrix of the proposal distribution
    
    :return: array with values of the generated markov chain aafter the burn-in period
    '''
    #We chose smart proposition parameters
    z0 = np.asarray([theta[0],theta[1]]+2*N*[0])
    
    diag_Sigma = [sig_t0, sig_v0] + N*[theta[3],theta[4]]
    Sigma = sigma*np.eye(z0.shape[0])*diag_Sigma
    z = np.asarray([z0])
    
    
    for i in range(burnIn+sample_size):
        u = np.random.uniform() #We sample uniform distribution
        z_prop = multivariate_normal(mean=z[-1], cov=Sigma).rvs(size=1) #Proposed value
        
        alpha = alpha_1(data, theta,z_prop, z[-1]) 
        
        if u<=alpha:
            z = np.append(z,[z_prop], axis=0)#We accept the proposed value
        else: 
            z = np.append(z,[z[-1]], axis=0) #We rejected 

    #We only return the samples after the burnin samples
    return z[burnIn+1:]
        
    
```

```python
def check_normal_sample(sampled, theo_mean, theo_st_dev, name):
    '''
    Compares, using a plot, the distribution of a set of points compared to the theoretical gaussian distribution 
    they are supposed to follow.
    :param sampled: array type variable with the generated sample
    :param theo_mean: mean of the gaussian distribution
    :param theo_st_dev: standard deviation of the gaussian distribution
    :param name: string type with the name of the variable for the title of the plot
    '''
    
    #Check symmetic random walk for the variable t_0
    mini = min(sampled)
    maxi = max(sampled)
    
    #Plot theoretical distribution
    x_array = np.linspace(mini, maxi, 100)
    y_array = np.asarray([norm.pdf(x, loc=theo_mean, scale=theo_st_dev) for x in x_array])

    #Plot distribution sampled by algorithm
    plt.plot(x_array, y_array, label="Posterior")
    plt.hist(sampled, bins=30, density=True, stacked=True, label="Posterior approximation (MH)")
    plt.title("Sampled VS theoretical distribution of "+ name)
    plt.legend()
    plt.show()
```

$$
\frac{\partial\mathcal{L}(y,z,\theta)}{\partial \bar{t_0}} =\frac{S_4}{\sigma_{t_0}^2} - \frac{2\bar{t_0}}{2\sigma_{t_0}^2}+\frac{\bar{\bar{t_0}}}{s_{t_0}^2}-\frac{2\bar{\bar{t_0}}}{2s_{t_0}^2} \implies \bar{t_0^*} = \left(\frac{1}{\sigma_{t_0}^2}+\frac{1}{s_t0^2}\right)^{-1}\left(\frac{S_4}{\sigma_{t_0}^2}+\frac{\bar{\bar{t_0}}}{s_{t_0}^2}\right)
$$

$$
\frac{\partial\mathcal{L}(y,z,\theta)}{\partial \sigma} = S_1\frac{NK2\sigma}{2\sigma^4}-\frac{NK}{\sigma}+ \frac{v^22\sigma}{2\sigma^4}-\frac{m+2}{\sigma} \implies {\sigma^*}^2 = \frac{S_1NK+v^2}{NK+m+2}
$$

By symetry we obtain that:

$$
\bar{v_0^*} = \left(\frac{1}{\sigma_{v_0}^2}+\frac{1}{s_{v_0}^2}\right)^{-1}\left(\frac{S_5}{\sigma_{v_0}^2}+\frac{\bar{\bar{v_0}}}{s_{v_0}^2}\right)
$$

$$
{\sigma_\xi^*}^2 = \frac{S_2N+v^2}{N+m+2}
$$

$$
{\sigma_\tau^*}^2 = \frac{S_3N+v^2}{N+m+2}

```python

def step_size(k, burnin = 100 , alpha = 0.5):
    '''
    Step size for the HM-SAEM as proposed in the homework.
    
    :param k: step of the HM-SAEM algorithm
    :param burnin: burn-in parameter
    :param alpha: must be greater or equal to 1/2 and smaller than 1
    '''
    if k<=burnin: return 1
    else: return (k-burnin)**(-alpha)
```

```python
def stoch_approx(data, z):
    '''
    Computes the term S(y,z^(k+1)) of the stochastic approximation step in MCMC-SAEM.
    
    :param data: y values
    :param z: z^(k+1)
    '''
    S = np.zeros([5])    
    N = data.shape[0] #number of individuals
    K = data.shape[1] #number of measurements per individual
                
    z_pop = z[0:2] #fixed effects 
    t_0 = z_pop[0]
    v_0 = z_pop[1]
                 
    z_re = z[2:] #random effects
    n = int(z_re.shape[0]/2 )
    xi = np.asarray([z_re[2*k] for k in range(n)]) #xi
    tau = np.asarray([z_re[2*k+1] for k in range(n)]) #tau 
      
    
    S = np.zeros(5)
    for i in range(N):
        for j in range(K):
            S[0] = S[0] + (data[i,j]-(p_0 + v_0*np.exp(xi[i])*(time[j]-t_0-tau[i])))**2
    S[0]=S[0]/(K*N)                 
    S[1]=(1/N)*((xi**2).sum())                         
    S[2] = (1/N)*((tau**2).sum())
    S[3] = t_0
    S[4] = v_0
            
                 
    return S
```

```python
def maximization(s_next, N, K ):
    '''
    Computes the maximization step of the MCMC-SAEM algorithm for curved exponential families, using the formulas
    found above for maximizing values.
    '''
    
    theta = np.zeros([5])
    
    #Approximation of \bar(t_0)
    t0 = s_next[3]
    theta[0] = (1/((1/sig_t0**2)+(1/s_t0**2)))*((t0/(sig_t0**2))+(t_bb_0/(s_t0**2)))
    
    #Approximation of \bar(v_0)
    v0 = s_next[4]
    theta[1] = (1/((1/sig_v0**2)+(1/s_v0**2)))*((v0/sig_v0**2)+(v_bb_0/s_v0**2))
    
    #Approximation of \sigma
    theta[2] = np.sqrt((s_next[0]*K*N+v**2)/(N*K+m+2))
    
    #Approximation of \sigma_psi
    theta[3] = np.sqrt((s_next[1]*N+v**2)/(N+m+2))
    
    #Approximation of \sigma_tau
    theta[4] = np.sqrt((s_next[2]*N+v**2)/(N+m+2))
    
    return theta
```

```python

def HM_SAEM(data, theta0,z0,alpha, maxIter=1000, burnInHM = 10000):
    '''
    Simulated E-M algorithm suing Metropolis Hastings in the simulation step.
    
    :return z_next: last sample of z
    :return theta_next: estimated value of the parameters
    :return evol_theta: array with all the values given to theta throughout the algorithm
    '''
    z_next = z0
    s_next = np.zeros(5)
    theta_next = theta0
    N = data.shape[0]
    K = data.shape[1]
    evol_theta=np.asarray([theta0])
    
    for i in range(maxIter):
        if i%10==0: print('SEAM iteration #', i)
        #Simulation
        z_prev=z_next
        z_next = SRW_hasting_metropolis_z(data,z_prev, theta_next, sample_size = 1, burnIn =burnInHM, sigma=0.00001)[-1]
        #Stochastic approximation
        s_prev=s_next
        s_next=s_prev + step_size(i, burnin = maxIter/2)*(stoch_approx(data,z_next)-s_prev)
        
        #Maximization
        theta_prev = theta_next
        theta_next = maximization(s_next, N, K )
        evol_theta=np.append(evol_theta,[theta_next], axis=0)
        
    return z_next, theta_next, evol_theta
```

### Question 5.
If we denote $\pi$ the a posterior distribution if $z_i$, it follows:

$$
\pi(\xi_i|\tau_i^{(k)}) = q(\xi_i|\tau_i^{(k)},y,z_{pop},\theta) = \frac{ q(y|\xi_i,\tau_i^{(k)},y,z_{pop},\theta)q(\xi_i|\tau_i^{(k)},y,z_{pop},\theta)}{q(y|\tau_i^{(k)},y,z_{pop},\theta)} \\
\propto q(y|\xi_i,\tau_i^{(k)},y,z_pop,\theta)q(\xi_i|\tau_i^{(k)}y,z_{pop},\theta) =
\mathcal{N}(\xi_i;0,\sigma_\xi)\prod^K_{j=1}\mathcal{N}(y_{ij};d(t_{ij}),\sigma)
$$
Where it is worth noting that $d(t_{ij})$ depends on both, $\xi_i$ and $\tau_i^{(k)}$, and that we only need to know the distribution up to a multiplication by a constant for the MH algorithm.

Analogously, 
$$
\pi(\tau_i|\xi_i^{(k+1)})
\propto q(y|\tau_i,\xi_i^{(k+1)},y,z_pop,\theta)q(\tau_i|\xi_i^{(k+1)},y,z_{pop},\theta) =
\mathcal{N}(\tau_i;0,\sigma_\tau)\prod^K_{j=1}\mathcal{N}(y_{ij};d(t_{ij}),\sigma)
$$

```python
def aux_alpha_xi(data, theta, xi, zpop, zi, i):
    K = data.shape[1]
    prod = 1
    
    tau = zi[2*i+1]
    t_0 = zpop[0]
    v_0 = zpop[1]
    
    for j in range(K):
        d_ij = p_0+ v_0*np.exp(xi)*(time[j]-t_0-tau)
        prod = prod*norm.pdf(data[i][j], loc=d_ij , scale=theta[2])
        
    prod = prod*norm.pdf(xi, loc=0 , scale=theta[3])
    
    return prod
def alpha_xi(data, theta,xi_prop, xi, zpop,zi, i):
    '''
    
    :param data: y_ij values
    :param theta: parameters of the model(\bar{t0},\bar{v0},\sigma,\sigma_xi,\sigma_tau)
    :param xi_prop: value proposed by the MH algorithm for xi[i]
    :param xi: current state of the markov chain xi[i]
    :param tau: parameter tau[i] corresponding to the same individual as the xi being sampled
    :param i: corresponding individual to the xi being sampled
    '''
    
    return aux_alpha_xi(data, theta,xi_prop, zpop,zi, i)/aux_alpha_xi(data, theta,xi, zpop,zi, i)
```

```python
def aux_alpha_tau(data, theta, tau, zpop,zi, i):
    K = data.shape[1]
    N = data.shape[0]
    prod = 1
    
    xi = zi[2*i]
    t_0 = zpop[0]
    v_0 = zpop[1]
    
    for j in range(K):
        d_ij = p_0+ v_0*np.exp(xi)*(time[j]-t_0-tau)
        prod = prod*norm.pdf(data[i][j], loc=d_ij , scale=theta[2])
        
    prod = prod*norm.pdf(tau, loc=0 , scale=theta[3])
    
    return prod

def alpha_tau(data, theta,tau_prop, tau, zpop,zi, i):
    
    return aux_alpha_tau(data, theta,tau_prop, zpop,zi, i)/aux_alpha_tau(data, theta,tau, zpop,zi, i)
    
```

```python
def SRW_hasting_metropolis_zi(data, zpop, theta,sample_size,burnIn = 1000, sigma=0.00001):
    '''
    Symmetric Random Walk Hasting Metropolis using multivariate gaussian distributions centered in the 
    the previous point as proposal distributions.
    
    :param z0: initial point of the Markov Chain
    :param maxIter: maximum number of iterations of the algorithm
    :param obj_func: objective funtion in which we want to create a sample
    :param Sigma: covariance matrix of the proposal distribution
    '''
    N = data.shape[0]
    z0 = np.zeros(2*N)
    zi_array = np.asarray([z0])
    
    for j in range(burnIn+sample_size):
        u = np.random.uniform() #We sample uniform distribution
        
        zi = zi_array[-1]

        for i in range(N):
            #Sample tau using HM sampling
            tau_i = zi[2*i+1]
            tau_i_prop = multivariate_normal(mean=tau_i, cov=sigma).rvs(size=1) #Proposed value
            alpha_tau_i = alpha_tau(data, theta,tau_i_prop, tau_i, zpop, zi,i)
            if u<=alpha_tau_i:
                zi[2*i+1] = tau_i_prop
            
            #Sample xi using HM sampling
            xi_i = zi[2*i]
            xi_i_prop = multivariate_normal(mean=xi_i, cov=sigma).rvs(size=1) #Proposed value
            alpha_xi_i = alpha_xi(data, theta,xi_i_prop, xi_i, zpop, zi,i)
            if u<=alpha_xi_i:
                zi[2*i] = xi_i_prop
      
        zi_array = np.append(zi_array,[zi],axis=0)

    #We only return the samples after the burnin samples
    return zi_array[burnIn:]
```

```python
def alpha_v0_aux(data, theta, v_0, zi,t_0):
    K = data.shape[1]
    N = data.shape[0]
    prod = 1
    
    tau = np.asarray([zi[2*j+1] for j in range(N)])
    xi = np.asarray([zi[2*j] for j in range(N)])

    for i in range(N):
        for j in range(K):
            d_ij = p_0+ v_0*np.exp(xi[i])*(time[j]-t_0-tau[i])
            prod = prod*norm.pdf(data[i][j], loc=d_ij , scale=theta[2])
        
    prod = prod*norm.pdf(v_0, loc =theta[1], scale = sig_v0)
    
    return prod

def alpha_v0(data, theta, v0_prop, v0, zi ,t_0):

    return alpha_v0_aux(data, theta, v0_prop, zi ,t_0)/alpha_v0_aux(data, theta, v0, zi ,t_0)
```

```python
def SRW_hasting_metropolis_zpop(data, zi, theta,sample_size,burnIn = 1000, sigma=0.001):
    '''
    Symmetric Random Walk Hasting Metropolis using multivariate gaussian distributions centered in the 
    the previous point as proposal distributions.
    
    :param z0: initial point of the Markov Chain
    :param maxIter: maximum number of iterations of the algorithm
    :param obj_func: objective funtion in which we want to create a sample
    :param Sigma: covariance matrix of the proposal distribution
    '''
    N = data.shape[0]

    t_b_0 = theta[0]
    v_b_0 = theta[1]
    
    zpop_array = np.asarray([[t_b_0,v_b_0]])

    for j in range(burnIn+sample_size):
        u = np.random.uniform() #We sample uniform distribution
        
        zpop = zpop_array[-1]

        #Sample t0
        t0 = zpop[0] 
        t0_prop =  multivariate_normal(mean = t_b_0,cov = sigma).rvs(size=1)
        alpha_t_0 = alpha_t0(data, theta, t0_prop, t0, zi ,zpop[1])
        if u<=alpha_t_0:
            zpop[0] = t0_prop
        
        #Sample v0
        v0 = zpop[1] 
        v0_prop =  norm(v_b_0,sigma).rvs(size=1)
        alpha_v_0 = alpha_v0(data, theta, v0_prop, v0, zi ,zpop[0])
        if u<=alpha_v_0:
            zpop[1] = v0_prop
      
        zpop_array = np.append(zpop_array,[zpop],axis=0)

    #We only return the samples after the burnin samples
    return zpop_array[burnIn:]
```

Consider a measurable and bounded function $h$. Then if we consider $X_n = x$ known and assume $f(\epsilon =0)=0$, we have:

$$
\mathbb{E}[h(Y)] = \mathbb{E}[h(Y)|\mathcal{B}=1]\mathbb{P}(\mathcal{B}=1)+\mathbb{E}[h(Y)|\mathcal{B}=0]\mathbb{P}(\mathcal{B}=0) = \frac{1}{2}\mathbb{E}[h(\epsilon x)]+\frac{1}{2}\mathbb{E}[h(\frac{x}{\epsilon})] \\
= \frac{1}{2}\int^1_{-1}h(\epsilon x)f(\epsilon)d\epsilon+\frac{1}{2}\int^0_{-1}h(\frac{x}{\epsilon})f(\epsilon)d\epsilon+\frac{1}{2}\int^1_{0}h(\frac{x}{\epsilon})f(\epsilon)d\epsilon.
$$

In the first integral we consider the change of variables  given by: 

$$y=\epsilon x \implies \epsilon = \frac{y}{x}, d\epsilon = \frac{dy}{x}$$.

In in the second and third integrals we use the change of variables 

$$y=\frac{x}{\epsilon} \implies \epsilon = \frac{x}{y}, d\epsilon = -\frac{dy}{x^2}$$. 

For the second change of variables we must take inot account the signs of $\epsilon$ and $x$, so we see two cases separetely depending on the sign of $x$.

If $x>0$, then, in the second integral, since $\epsilon<0$, the upper bound of the integral becomes $-\infty$ after the change of variables. Analogously, for the third integral, the lower bound is $+\infty$ since $\epsilon>0$. Hence, if $x>0$


$$
\mathbb{E}[h(Y)] = \frac{1}{2}\int^x_{-x}h(y)f(\frac{y}{x})\frac{1}{x}dy+\frac{1}{2}\int_{-\infty}^{-x}h(y)f(\frac{x}{y})\frac{x}{y^2}dy+\frac{1}{2}\int_x^{\infty}h(y)f(\frac{x}{y})\frac{x}{y^2}dy.
$$

Analogously, in the case of $x<0$

$$
\mathbb{E}[h(Y)] = -\frac{1}{2}\int_x^{-x}h(y)f(\frac{y}{x})\frac{1}{x}dy-\frac{1}{2}\int^{+\infty}_{-x}h(y)f(\frac{x}{y})\frac{x}{y^2}dy-\frac{1}{2}\int_{-\infty}^{x}h(y)f(\frac{x}{y})\frac{x}{y^2}dy.
$$

Then, using the fact that the sign of $x$ is equal to $\frac{x}{||x||}$, and that, if $x<0$, $-|x|=x$ and if $x>0$, $|x| = x$ we, the expression in both scenarios can be rewritten as 

$$
\mathbb{E}[h(Y)] = \frac{x}{||x||}\frac{1}{2}\int^x_{-x}h(y)f(\frac{y}{x})\frac{1}{x}dy+\frac{x}{||x||}\frac{1}{2}\int_{-\infty}^{-|x|}h(y)f(\frac{x}{y})\frac{x}{y^2}dy+\frac{x}{||x||}\frac{1}{2}\int_{|x|}^{\infty}h(y)f(\frac{x}{y})\frac{x}{y^2}dy.
$$

In conclusion, the jumping distribution is given by

$$
q(x,y) = \frac{1}{2|x|}h(y)f(\frac{y}{x})\mathbb{1}_{|y|<|x|}+\frac{|x|}{2}h(y)f(\frac{x}{y})\frac{1}{y^2}\mathbb{1}_{y<-|x|}+\frac{|x|}{2}h(y)f(\frac{x}{y})\frac{1}{y^2}\mathbb{1}_{y>|x|}= \\

```python
from scipy.stats import expon
def multiplicative_HM_exp(x0,sample_size, burnIn = 10000):
    '''
    Multiplicative Hasting Metropolis algorithm for an exponential distribution using a 
    uniform distribution that avoids 0 on (-1,1) as the f.
    '''
    
    x = np.asarray([x0])
    
    
    for i in range(burnIn+sample_size):
        B = np.random.binomial(1, 1/2)
        eps = 0
        
        #We avoid epsilon to be 0
        while eps == 0:
            eps = np.random.uniform(-1, high=1)
        
        if B==1: x_prop = eps*x[-1]
        else: x_prop=x[-1]/eps
 
        alpha =  (expon.pdf(x_prop)*abs(x_prop))/(expon.pdf(x[-1])*abs(x[-1]))
        
        u = np.random.uniform()
        if u<=alpha:
            x = np.append(x,[x_prop], axis=0)#We accept the proposed value
        else: 
            x = np.append(x,[x[-1]], axis=0) #We rejected 

    #We only return the samples after the burnin 
    return x[burnIn+1:]
        
    
    
```

```python
def Gibbs_Augmentation(z0, maxIters=1000):
    '''
    Gibbs algorithm using data augmentation.
    
    '''
    
    sample = np.asarray([z0])
    
    for i in range(maxIters):
        
        z = sample[-1]
        z[0] = norm.rvs(0, 1/np.sqrt(z[1]))
        
        z[1] = np.random.gamma(shape = 5/2, scale = 1/((z[0]**2/2)+2))
                
        sample = np.append(sample, [z], axis =0)

    return sample
```
